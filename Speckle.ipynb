```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc 
```

```python
imagen=misc.imread("Imagen_gris6.jpg")
print(imagen.shape)
```

```python
matrix=np.asarray(imagen)
m=2026
l=1971
n=m*l
```

```python
data=[0]*n
k=0

for i in range(0,m):
    for j in range(0,l):
        data[k]=matrix[i][j] #convierto los datos de la matriz en un arreglo
        k+=1 #contador
        
        
M=np.mean(data)
De=np.std(data)
s=np.random.normal(M,De,n)

plt.figure()
plt.title("Imagen # n")
plt.xlabel("Valor en escala gris")
plt.ylabel("Numero de repeticiones")
plt.hist(s,200,normed=True)
plt.hist(data,200,normed=True)
plt.show()

print("Media =",M,"Desviacion estandar=",De)
```

```python
k=0
for i in range(0,n,2): #para hallar n/2
    k+=1

data1=[0]*k
k=0
j=0
for i in range(0,n,2):
    data1[k]=data[j]+data[j+1]
    
    k+=1
    j+=2
    

    
M1=np.mean(data1)
De1=np.std(data1)
s=np.random.normal(M1,De1,k)

plt.figure()
plt.title("Imagen # n/2")
plt.xlabel("Valor en escala gris")
plt.ylabel("Numero de repeticiones")
plt.hist(s,200,normed=True)
plt.hist(data1,200,normed=True)
plt.show()
print("Media =",M1,"Desviacion estandar=",De1)
```
